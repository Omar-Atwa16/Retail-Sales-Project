```python
# Complete Data Insights

import pandas as pd

df = pd.read_csv(r'/content/synthetic_retail_sales.csv')
print(f"The Full Data:\n{df}")

print ("\n\n")

numberOfTransactions = df['TransactionID'].count()
print(f"Number of Transactions: {numberOfTransactions}")

print ("\n\n")

totalSales = df['TotalSale'].sum().round(2)
print(f"Total Sales: ${totalSales}")

print ("\n\n")

totalUnitsSold = df['QuantitySold'].sum()
print(f"Total Units Sold: {totalUnitsSold}")

print ("\n\n")

unitsSoldPerCat = df.groupby('ProductCategory')['QuantitySold'].sum().sort_values(ascending=False).reset_index(name = 'UnitsSold')
print(f"Units Sold Per Product Category:\n{unitsSoldPerCat}")

print ("\n\n")

unitsSoldPerCity = df.groupby('StoreLocation')['QuantitySold'].sum().sort_values(ascending=False).reset_index(name = 'UnitsSold')
print(f"Units Sold Per City:\n{unitsSoldPerCity}")

print ("\n\n")

salesPerCat = df.groupby('ProductCategory')['TotalSale'].sum().round(2).sort_values(ascending=False).reset_index(name = 'TotalSales')
print(f"Total Sales Per Product Category:\n{salesPerCat}")

print ("\n\n")

salesPerCity = df.groupby('StoreLocation')['TotalSale'].sum().round(2).sort_values(ascending=False).reset_index(name = 'TotalSales')
print(f"Total Sales Per City:\n{salesPerCity}")

print ("\n\n")

avgAgePerCat = df.groupby('ProductCategory')['CustomerAge'].mean().round().sort_values(ascending=False).reset_index(name = 'AverageAge')
print(f"Average Customer Age Per Product Category:\n{avgAgePerCat}")

print ("\n\n")

paymentMethodCount = df['PaymentMethod'].value_counts().reset_index(name='UsageCount')
print(f"Payment Method Usage Counter:\n{paymentMethodCount}")

print ("\n\n")

amountPerPaymentMethod = df.groupby('PaymentMethod')['TotalSale'].sum().round(2).sort_values(ascending=False).reset_index(name = 'Amount')
print(f"Amount Per Payment Method:\n{amountPerPaymentMethod}")

print ("\n\n")

refundCount = (df['Returned'] == 'Yes').sum()
print(f"Number of Refunds: {refundCount}")

print ("\n")

refunded_amount = df[df['Returned'] == 'Yes']['TotalSale'].sum().round(2)
print(f"Total Refunded Amount: ${refunded_amount}")

print ("\n")

salesAfterRefund = (totalSales - refunded_amount).round(2)
print(f"Total Sales After Refunds:${salesAfterRefund}")
```

```python
# Model To Predict The Payment Method Based On Product Category And The Total Amount With Complete Data

import pandas as pd
import numpy as np
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split

df = pd.read_csv(r'/content/synthetic_retail_sales.csv')

x = df[['ProductCategory', 'TotalSale']]
y = df['PaymentMethod']

X = pd.get_dummies(x, columns=['ProductCategory'])
y = pd.get_dummies(y)

x_train, x_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=16)

knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(x_train, y_train)

prediction = knn.predict(x_test)

predicted_labels = [y.columns[np.argmax(p)] for p in prediction]

for i, pred in enumerate(predicted_labels):
    print(f"Transaction {i+1}: Predicted Payment Method: {pred}")

print ("\n\n")

predicted_series = pd.Series(predicted_labels)
predictions_counter = predicted_series.value_counts()

print(f"Number of Predictions For Each payment Method:\n{predictions_counter}")
```

```python
# Model To Predict The Payment Method Based On Product Category And The Total Amount After Handling Missing Data

import pandas as pd
import numpy as np
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split

df = pd.read_csv(r'/content/synthetic_retail_sales_hidden.csv')

df.fillna(df.mean(numeric_only = True), inplace=True)
df.fillna(df.mode().iloc[0], inplace=True)


x = df[['ProductCategory', 'TotalSale']]
y = df['PaymentMethod']

X = pd.get_dummies(x, columns=['ProductCategory'])
y = pd.get_dummies(y)

x_train, x_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=16)

knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(x_train, y_train)

prediction = knn.predict(x_test)

predicted_labels = [y.columns[np.argmax(p)] for p in prediction]

for i, pred in enumerate(predicted_labels):
    print(f"Transaction {i+1}: Predicted Payment Method: {pred}")

print ("\n\n")

predicted_series = pd.Series(predicted_labels)
predictions_counter = predicted_series.value_counts()

print(f"Number of Predictions For Each payment Method:\n{predictions_counter}")
```
